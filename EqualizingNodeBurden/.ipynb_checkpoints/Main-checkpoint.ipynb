```python
#import
import node
import blockchain
import block
import access_pattern
import time

# 設定値
node_num = 100
run_time = 500
genInterval = 10 #生成間隔(s)
```

```python
start_time = time.time()


#ノードの初期化
nodes = []
for i in range(node_num):
    nodes.append(node.node("122.2.2.2",i,50))
    print("node ", i, "is :", nodes[i].node_ID)
    if i != 0:
        nodes[i].join(nodes[i-1])

#BlockChainの初期化
bc = blockchain.blockchain()

#Kademliaノードの経路表構築
#Kとαの指定



i = 0
while i < run_time :#１ループ１分とする（Ethereumなら、生成間隔は１２秒くらいなため、１秒とする）
#     print(time.time())
    if i % genInterval == 0: #１０単位時間経ったら
        newBlock = bc.createBlock(i)
#         print("Now is ", i, "s")
        for n in nodes:
            n.saveNewBlock(newBlock)
    
    access_pattern.access_pattern().Access(nodes, i)
    i += 1

    
# 生成終了
# 確認（データ永続性）

print("データ永続性を確認します")
for i in range(run_time % genInterval):
    key = bc.hash(bc.chain[i].header)
    result = nodes[0].lookup_value(key)
    if result == None:
        print(key, "は、失われました")
#     else:
#         print()


# for n in nodes:
#     print("------------------------\nThe Demanded Table in Node ", n.node_ID, " when the Chain length is ", len(n.Headers))
#     for i in range(len(n.Demanded)):
#         print("ID: ", i,  ", Body: ", n.Demanded[i]["body"], ", headerIndex: ", n.Demanded[i]["headerIndex"])

# for n in nodes:
#     print("------------------------\nThe Non-Demanded Table in Node ", n.node_ID)
#     print(n.print_index())
    
    
# for n in nodes:
#     print("ノード", n.node_ID, "が各ブロックに近いK個のノードを示します")
#     for b in bc.chain:
#         k_list = n.lookup(n.hash(n.hash(b.header)))
#         print("ブロック: ", n.hash(b.header), "(", n.hash(n.hash(b.header)),")", "のK_listは、",k_list, "です")


exe_time = time.time() -start_time
print("")
print(exe_time)
```
